## 🔄 Cell 5 — Free-form Chat (loop)
Run this cell to have a back-and-forth conversation in the notebook output.

```python
# List of questions to ask in sequence (simulates a conversation)
questions = [
    "Which flats have not paid for March 2025?",
    "What about April 2025?",           # tests memory (no month context needed)
    "Who is the resident of flat 305?",
    "How much did they pay in total?",   # tests memory (refers to flat 305)
]

session_id = 'loop-session'
for q in questions:
    print(f'\n👤 You : {q}')
    ans = chat(q, thread_id=session_id)
    print(f'🤖 Bot : {ans}')
    print('-' * 50)
```

## 🔁 Cell 6 — Start a fresh conversation (reset memory)

```python
import time

# Use a unique thread_id to start a brand new conversation with no history
new_session = f'fresh-{int(time.time())}'
print(f'New session ID: {new_session}')

response = chat("What is the net balance for 2025?", thread_id=new_session)
print(response)
```
